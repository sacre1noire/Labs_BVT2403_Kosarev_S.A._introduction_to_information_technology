```python
class Vehicle:
    def __init__(self,make,model):
        self.make=make
        self.model=model
    def get_info(self):
        return f'Название марки {self.make}, модель: {self.model}'
class Car(Vehicle):
    def __init__(self,make,model,fuel_type):
        super().__init__(make,model)
        self.ft=fuel_type
    def get_info(self):
        s=super().get_info()
        return s +f', тип топлива: {self.ft}'
auto=Vehicle('Nissan','Cefiro')
print(auto.get_info())
add=Car('Nissan','Silvia','Бензин - 98')
print(add.get_info())
```
