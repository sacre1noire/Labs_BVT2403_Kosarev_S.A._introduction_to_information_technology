```python
class Book():
    def __init__(self,title,author,year):
        self.title=title
        self.author=author
        self.year=year
    def get_info(self):
        print(f'Название книги: {self.title}; Автор: {self.author}; Год издания: {self.year}')
book1=Book("Богатый папа, бедный папа","Роберт Кийосаки и Шэрон Л Лектер", 1997)
book1.get_info()
```

```python
class Circle():
    def __init__(self,rad):
        self._rad=rad
    @property
    def rad(self):
        return self._rad
    @rad.setter
    def rad(self,new_rad):
        self._rad=new_rad
kryg=Circle(25)
kryg.get_radius()
kryg.rad=767
print(kryg.get_radius())
```
